```python
import cv2 as cv
import numpy as numpy
import matplotlib.pyplot as plt
```

```python
gbr1=cv.imread("Assets/citra1.PNG")
gbr2=cv.imread("Assets/citra2.PNG")

plt.subplot(2,2,1)
plt.imshow(gbr1)
plt.axis('off')
plt.title("citra1")

plt.subplot(2,2,2)
plt.imshow(gbr2)
plt.axis('off')
plt.title("citra2")
```

```python
plt.subplot(2,2,1)
plt.imshow(gbr1, cmap='gray')
plt.axis('off')
plt.title("Gambar 1")

plt.subplot(2,2,2)
plt.imshow(gbr2, cmap='gray')
plt.axis('off')
plt.title("Gambar 2")
```
